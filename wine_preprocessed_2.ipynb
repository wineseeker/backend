```python
import pandas as pd

file_path = '/content/drive/MyDrive/wine_preprocessed.csv'

df = pd.read_csv(file_path)
```

```python
# wine_id가 3874052와 87832인 데이터 제거
df = df[(df['wine_id'] != 3874052) & (df['wine_id'] != 87832)]
```

```python
df_rating_average = df.copy()

df_rating_average
```

```python
# rating_average 열에서 이상치 식별
outliers = df_rating_average[(df_rating_average['rating_average'] <= 0) | (df_rating_average['rating_average'] >= 5)]

print(outliers)
```

```python
# 이상치 제거
df_rating_average = df_rating_average.drop(outliers.index)

# 이상치가 제거된 데이터프레임 확인
print(df_rating_average)
```

```python
df_alcohol = df_rating_average.copy()

df_alcohol
```

```python
# alcohol 열의 이상치 식별
Q1 = df_alcohol['alcohol'].quantile(0.25)
Q3 = df_alcohol['alcohol'].quantile(0.75)
IQR = Q3 - Q1

# 이상치 경계 계산
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 이상치 식별
outliers = df_alcohol[(df_alcohol['alcohol'] < lower_bound) | (df_alcohol['alcohol'] > upper_bound)]

# 이상치 출력
print("Outliers\n")
print(outliers)
```

```python
df_alcohol = df_alcohol.drop(outliers.index)

# 이상치가 제거된 데이터프레임 확인
print(df_alcohol)
```

```python
df_winery_ratings_average = df_alcohol.copy()

df_winery_ratings_average
```

```python
# 'winery_ratings_average' 열의 이상치 식별
outliers = df_winery_ratings_average[(df_winery_ratings_average['winery_ratings_average'] <= 0) | (df_winery_ratings_average['winery_ratings_average'] > 5)]

# 이상치 출력
print("Winery Ratings Average 열의 이상치:\n")
print(outliers[['wine_id', 'winery_ratings_average']])
```

```python
df_winery_ratings_average = df_winery_ratings_average.drop(outliers.index)

# 이상치가 제거된 데이터프레임 확인
print(df_winery_ratings_average)
```

```python
df_intensity = df_winery_ratings_average.copy()

df_intensity
```

```python
# 'intensity' 열의 이상치 식별
outliers = df_intensity[(df_intensity['intensity'] <= 0) | (df_intensity['intensity'] > 5)]

# 이상치 출력
print("intensity 열의 이상치:\n")
print(outliers[['wine_id', 'intensity']])
```

```python
df_intensity = df_intensity.drop(outliers.index)

# 이상치가 제거된 데이터프레임 확인
print(df_intensity)
```

```python
df_sweetness = df_intensity.copy()

df_sweetness
```

```python
# 'sweetness' 열의 이상치 식별
outliers = df_sweetness[(df_sweetness['sweetness'] <= 0) | (df_sweetness['sweetness'] > 5)]

# 이상치 출력
print("sweetness 열의 이상치:\n")
print(outliers[['wine_id', 'sweetness']])
```

```python
df_tannin = df_sweetness.copy()

df_tannin
```

```python
# 'tannin' 열의 이상치 식별
outliers = df_tannin[(df_tannin['tannin'] <= 0) | (df_tannin['tannin'] > 5)]

# 이상치 출력
print("tannin 열의 이상치:\n")
print(outliers[['wine_id', 'tannin']])
```

```python
df_tannin = df_tannin.drop(outliers.index)

# 이상치가 제거된 데이터프레임 확인
print(df_tannin)
```
